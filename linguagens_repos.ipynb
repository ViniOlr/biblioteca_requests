Agora iremos fazer uma requisição para a API do GitHub, e realizar a paginação para acessar todos os repositórios.

```python
# Fazendo um for, para percorrer de 1 a 6, e com base nesse número, iremos passar como parâmetro para recuperar os repositórios de cada página
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

# Transformando os dados

Recuperando os nomes dos repositórios

```python
repos_name = []

# Percorrendo cada página dos repositórios
for page in repos_list:
    # Percorrendo cada repositório da página
    for repo in page:
        # Armazenando o nome do repositório
        repos_name.append(repo.get('name'))

print('Qtd. de repositórios: ' + str(len(repos_name)))
# Exibindo os primeiros 10 primeiros
repos_name[:10]
```

Recuperando as linguagens de programação dos repositórios.

```python
repos_language = []

# Percorrendo cada página dos repositórios
for page in repos_list:
    # Percorrendo cada repositório da página
    for repo in page:
        # Armazenando a linguagem de programação do repositório
        repos_language.append(repo.get('language'))

# Exibindo os primeiros 10 primeiros
repos_language[:10]
```

Criando um DataFrame com base nos dados recuperados

```python
# Criando um DataFrame
dados_amz = pd.DataFrame()

# Definindo o nome da coluna do DataFrame e atribuindo valores a ela
dados_amz['repository_name'] = repos_name
dados_amz['repository_language'] = repos_language

dados_amz
```

Salvando o DataFrame no formato de csv

```python
nome_arquivo = 'amazon.csv'
dados_amz.to_csv(nome_arquivo)
```

# Armazenando os dados

Criando repositório via requisição do tipo POST

```python
url = 'https://api.github.com/user/repos'

data = {
    'name':'linguagens-utilizadas',
    'description': 'Repositórios com as linguagens dos repositórios da Amazon (Repositório, criado por requisição da API do GitHub, de exemplo durante o curso da Alura "Python e APIS")',
    'private':False
}

response = requests.post(url, headers=headers, json=data)
response.status_code
```

Formatando o arquivo para subir no repositório criado. </br>
Transformar em formato binário -> transformar em base 64

```python
# Lendo o arquivo em formato binário
with open(nome_arquivo, 'rb') as f:
    file_content = f.read()
    f.close()

# Transformando em base 64
encoded_content = base64.b64encode(file_content)
```

Fazendo uploado do arquivo com, para o repositório criado, com put

```python
url_base = 'https://api.github.com'
username = 'ViniOlr'
repo = data.get('name')
path = nome_arquivo # Caminho, dentro do nosso repositório, onde será salvo o arquivo

url = f'{url_base}/repos/{username}/{repo}/contents/{path}'

# Definindo informações de upload
data = {
    'message': 'Adicionando um novo arquivo',
    'content': encoded_content.decode('utf-8') # transforma o dado de base 64, para a API entender que é um texto
}

response = requests.put(url, headers=headers, json=data)
response.status_code
```
